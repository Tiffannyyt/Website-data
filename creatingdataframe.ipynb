```python
import pandas as pd
city_names = pd.Series([
    'Atlanta',
    'Austin',
    'Kansas City',
    'New York City',
    'Portland',
    'San Fransisco',
    'Seattle',
])

population = pd.Series([
    498044,
    964254,
    491918,
    8398748,
    653115,
    883305,
    744955,
])

num_airports = pd.Series([
    2,
    2,
    8,
    3,
    1,
    3,
    2,
])

df = pd.DataFrame({
    'City Name': city_names,
    'Population': population,
    'Airports': num_airports,
})

print(df)

df.describe(include='all')
df.head(12)
df.sample(5) #ngambil 5 random data 

#accessing values 
df['City Name']

df.iloc[3]
df.iloc[1:3]
df.iloc[[1,3]]

df.loc[2, 'City Name']
df.loc[[2, 3], ['City Name', 'Airports']]

df.loc[6, 'Airports'] = 8
df

#modifying entire column
df["Population"] /= 10000
df

#assigning to a new column name 
df['Population_M'] = df['Population'] / 1000
df

```
